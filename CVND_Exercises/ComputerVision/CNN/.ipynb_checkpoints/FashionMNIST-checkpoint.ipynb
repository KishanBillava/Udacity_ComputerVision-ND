```python
from torchvision.datasets import FashionMNIST
from torch.utils.data import DataLoader
from torchvision import transforms
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
import torch.optim as optim
```

```python
data_transform = transforms.ToTensor()
```

```python
train_data = FashionMNIST(root='./data', train=True, 
                          download=True, transform=data_transform)

test_data = FashionMNIST(root='./data', train=False, 
                         download=True, transform=data_transform)
```

```python
print('Train data, number of images: ', len(train_data))
print('Test data, number of images: ', len(test_data))
```

```python
# batch_size affects the loss
batch_size = 20
```

```python
train_loader = DataLoader(train_data, 
                          batch_size=batch_size, shuffle=True)

test_loader = DataLoader(test_data, 
                         batch_size=batch_size, shuffle=True)
```

```python
# Image Classes --> 10
classes = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
           'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot']
```

```python
# Visualize some training data

dataiter = iter(train_loader)
images, labels = dataiter.next()
images = images.numpy()

# plot
fig = plt.figure(figsize=(10,10))
for idx in np.arange(batch_size):
    ax = fig.add_subplot(4,batch_size/4, idx+1, 
                         xticks=[], yticks=[])
    ax.imshow(np.squeeze(images[idx]), cmap='gray')
    ax.set_title(classes[labels[idx]])
```

```python
# takes in a 1 channel (grayscale) image
# outputs 10 feature maps as output, 
# after convolving the image with 3x3 filters.
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        # 1x(28x28) --> 10x(3x3) --> 10x26x26
        self.conv1 = nn.Conv2d(1,10,3)
        # 10x26x26 --> 2x2 -->  10x(13x13)
        self.pool = nn.MaxPool2d(2,2)
        
        # 10x(13x13) --> 20x(3x3) --> 20x(11x11)
        self.conv2 = nn.Conv2d(10, 20, 3)
        # 20x(11x11) -->  2x2 --> 20x(5x5)
        
        # 10 output channels (for 10 classes)
        self.fc1 = nn.Linear(20*5*5, 10)
        
    def forward(self, x):
        # two conv--relu  + pool layers
        # x input ---> conv & relu --> pool
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        
        # flatten the inputs into a vector 
        x = x.view(x.size(0), -1)
        
        # one linear layer
        x = self.fc1(x)
        # softmax layer to convert the 10 outputs into a distribution of class scores
        x = F.log_softmax(x, dim=1)
        
        # final output 
        return x 
```

```python
# print your Net
net = Net()
print(net)
```

```python
# loss funcation 
criterion = nn.NLLLoss()
# and optimizer ---> stochastic gradient descent with a small learning rate
optimizer = optim.SGD(net.parameters(), lr=0.001)
```
